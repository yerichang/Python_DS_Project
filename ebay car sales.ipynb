```python
brand_counts = autos['brand'].value_counts(normalize=True)
above_5percent = brand_counts[brand_counts>0.05].index
print(above_5percent)
```

select brands that have over 5 % of the total values

```python
mean_price = {}
mean_mileage = {}

for b in above_5percent:
    avg_price = autos.loc[autos["brand"] == b,"price"].mean()
    avg_mileage = autos.loc[autos["brand"] == b, 'odometer_km'].mean()
    mean_price[b] = round(avg_price,2)
    mean_mileage[b] = round(avg_mileage,2)
    
print(mean_price)
print(mean_mileage)
```

```python
mp_series = pd.Series(mean_price)
mm_series = pd.Series(mean_mileage)

mean_df = pd.DataFrame(mp_series, columns=['mean_price'])
mean_df["mean_mileage"] = mm_series

print(mean_df)
```

Mercedes Benz is the most expensive and Opel is the least expensive among top 6 brands but their mean mileages do not seem to correlate.
